```python
class RacetrackEnv:
    """
    - 0: Off-track (boundary)
    - 1: Track
    - 2: Starting line
    - 3: Finish line
    """

    def __init__(self, grid=None, noise_prob=0.1):
        self.grid = np.array(grid if grid is not None else ENV_1)

        self.noise_prob = noise_prob

        self.height, self.width = self.grid.shape

        self.start_positions = list(zip(*np.where(self.grid == 2)))
        self.finish_positions = set(zip(*np.where(self.grid == 3)))

        self.max_velocity = 4

        self.actions = [(dx, dy) for dx in [-1, 0, 1] for dy in [-1, 0, 1]]

        # State: (x, y, vx, vy)
        self.state = None

        self.reset()

    def reset(self):
        start_pos = self.start_positions[np.random.randint(len(self.start_positions))]

        self.state = (start_pos[0], start_pos[1], 0, 0)

        return self.state

    def _is_collision(self, y1, x1, y2, x2):
        dy = y2 - y1
        dx = x2 - x1
        steps = max(abs(dx), abs(dy))

        if steps == 0:
            return False

        for i in range(1, steps + 1):
            y = int(y1 + (dy * i) / steps)
            x = int(x1 + (dx * i) / steps)

            if 0 <= y < self.height and 0 <= x < self.width:
                if self.grid[y, x] == 0:
                    return True
            else:
                return True

        return False

    def step(self, action):
        """
        Args:
            action: Integer index (0-8) representing the action

        Returns:
            next_state: Tuple (y, x, vy, vx)
            reward: -1 for each step
            done: True if reached finish line
        """
        y, x, vy, vx = self.state

        delta_vy, delta_vx = self.actions[action]

        # Apply noise
        if np.random.random() < self.noise_prob:
            delta_vy, delta_vx = 0, 0

        new_vy = np.clip(vy + delta_vy, 0, self.max_velocity)
        new_vx = np.clip(vx + delta_vx, 0, self.max_velocity)

        if new_vy == 0 and new_vx == 0:
            new_vy, new_vx = vy, vx

        new_y = y - new_vy  # Subtract to move up
        new_x = x + new_vx  # Add to move right

        collision = self._is_collision(y, x, new_y, new_x)

        if collision:
            start_idx = np.random.randint(len(self.start_positions))
            start_pos = self.start_positions[start_idx]

            self.state = (start_pos[0], start_pos[1], 0, 0)

            return self.state, -1, False

        if self.grid[new_y, new_x] == 3:
            self.state = (new_y, new_x, new_vy, new_vx)

            return self.state, -1, True

        self.state = (new_y, new_x, new_vy, new_vx)

        return self.state, -1, False

    def render(self, trajectory=None):
        plt.figure(figsize=(12, 10))

        display_grid = self.grid.copy().astype(float)

        # Add blue color for trajectory cells
        cmap_colors = [
            "#777777",
            "white",
            "#e18272",
            "#82b680",
            "#4169e1",
        ]  # Added blue

        from matplotlib.colors import ListedColormap

        cmap = ListedColormap(cmap_colors)

        # Color trajectory cells blue
        if trajectory is not None:
            for pos in trajectory:
                display_grid[pos[0], pos[1]] = 4  # Set to 4 for blue color

            # Mark start and end positions differently if desired
            # Optionally, you can still add markers for start/end
            traj_y = [pos[0] for pos in trajectory]
            traj_x = [pos[1] for pos in trajectory]
            plt.plot(traj_x[0], traj_y[0], "ro", markersize=10, label="Start")
            plt.plot(traj_x[-1], traj_y[-1], "go", markersize=10, label="End")
            plt.legend()

        plt.imshow(display_grid, cmap=cmap, origin="upper")

        nrows, ncols = display_grid.shape

        for y in range(nrows + 1):
            plt.axhline(y - 0.5, color="black", linewidth=0.5)

        for x in range(ncols + 1):
            plt.axvline(x - 0.5, color="black", linewidth=0.5)

        plt.grid(False)
        plt.tight_layout()
        plt.show()
```

```python
class OffPolicyMC:
    def __init__(self, env, epsilon=0.1, gamma=0.9, init_value=0.0):
        self.env = env
        self.epsilon = epsilon
        self.gamma = gamma
        self.init_value = init_value

        self.Q = {}
        self.C = {}

        self.valid_positions = list(zip(*np.where((env.grid == 1) | (env.grid == 2))))

    def get_q_value(self, state, action):
        if (state, action) not in self.Q:
            self.Q[(state, action)] = self.init_value

            self.C[(state, action)] = 0

        return self.Q[(state, action)]

    def get_greedy_action(self, state):
        q_values = [self.get_q_value(state, a) for a in range(len(self.env.actions))]

        max_q = max(q_values)

        best_actions = [a for a in range(len(self.env.actions)) if q_values[a] == max_q]

        return np.random.choice(best_actions)

    def get_epsilon_greedy_action(self, state):
        if np.random.random() < self.epsilon:
            return np.random.randint(len(self.env.actions))
        else:
            return self.get_greedy_action(state)

    def generate_episode(self, use_exploring_starts=False):
        episode = []

        if use_exploring_starts:
            pos = self.valid_positions[np.random.randint(len(self.valid_positions))]
            vy = np.random.randint(0, self.env.max_velocity + 1)
            vx = np.random.randint(0, self.env.max_velocity + 1)

            if vy == 0 and vx == 0:
                if np.random.random() < 0.5:
                    vy = 1
                else:
                    vx = 1

            self.env.state = (pos[0], pos[1], vy, vx)

            state = self.env.state
        else:
            state = self.env.reset()

        done = False
        max_steps = 1000  # Prevent infinite loops
        steps = 0

        while not done and steps < max_steps:
            action = self.get_epsilon_greedy_action(state)
            next_state, reward, done = self.env.step(action)
            episode.append((state, action, reward))
            state = next_state
            steps += 1

        return episode

    def train(self, num_episodes=10000):
        episode_lengths = []

        for episode_num in range(num_episodes):
            use_exploring_starts = episode_num < num_episodes // 2

            episode = self.generate_episode(use_exploring_starts)
            episode_lengths.append(len(episode))

            G = 0
            W = 1

            for t in range(len(episode) - 1, -1, -1):
                state, action, reward = episode[t]

                G = self.gamma * G + reward

                self.C[(state, action)] = self.C.get((state, action), 0) + W

                if self.C[(state, action)] > 0:
                    prev_q = self.Q.get((state, action), self.init_value)
                    weight = W / self.C[(state, action)]

                    self.Q[(state, action)] = prev_q + weight * (G - prev_q)

                greedy_action = self.get_greedy_action(state)

                # If action taken is not the greedy action, break (W becomes 0)
                if action != greedy_action:
                    break

                # π(A_t|S_t) = 1 (greedy action has probability 1)
                # b(A_t|S_t) = (1-ε) + ε/|A| for greedy action, ε/|A| otherwise
                num_actions = len(self.env.actions)
                prob_behavior = (1 - self.epsilon) + self.epsilon / num_actions
                W = W * (1.0 / prob_behavior)

            if (episode_num + 1) % 10_000 == 0:
                avg_length = np.mean(episode_lengths[-10_000:])

                print(
                    f"Episode {episode_num + 1}/{num_episodes}, "
                    f"Avg Length (last 10,000): {avg_length:.2f}"
                )

        return episode_lengths

    def get_state_values(self):
        state_values = {}

        for (state, action), q_val in self.Q.items():
            if state not in state_values:
                state_values[state] = q_val
            else:
                state_values[state] = max(state_values[state], q_val)

        return state_values

    def visualize_value_function(self):
        value_grid = np.zeros((self.env.height, self.env.width))
        count_grid = np.zeros((self.env.height, self.env.width))

        state_values = self.get_state_values()

        for state, value in state_values.items():
            y, x, vy, vx = state

            if 0 <= y < self.env.height and 0 <= x < self.env.width:
                value_grid[y, x] += value
                count_grid[y, x] += 1

        if count_grid.any():
            avg_value_grid = np.where(count_grid > 0, value_grid / count_grid, 0)
        else:
            avg_value_grid = np.zeros_like(value_grid)

        masked_values = np.ma.masked_where(self.env.grid == 0, avg_value_grid)

        plt.figure(figsize=(14, 10))

        im = plt.imshow(masked_values, cmap="RdYlGn", origin="upper")
        plt.colorbar(im, label="Average State Value")

        for y in range(self.env.height + 1):
            plt.axhline(y - 0.5, color="black", linewidth=0.5)

        for x in range(self.env.width + 1):
            plt.axvline(x - 0.5, color="black", linewidth=0.5)

        plt.title("Average State Values for Each Position", fontsize=14)
        plt.grid(False)
        plt.tight_layout()
        plt.show()

    def test_policy(self, num_episodes=3):
        original_noise = self.env.noise_prob
        self.env.noise_prob = 0

        for ep in range(num_episodes):
            state = self.env.reset()

            trajectory = [(state[0], state[1])]

            done = False

            steps = 0
            max_steps = 1000

            while not done and steps < max_steps:
                action = self.get_greedy_action(state)

                state, reward, done = self.env.step(action)

                trajectory.append((state[0], state[1]))

                steps += 1

            print(f"Test Episode {ep + 1}: {steps} steps")

            self.env.render(trajectory)

        self.env.noise_prob = original_noise
```

```python
env = RacetrackEnv(grid=ENV_1, noise_prob=0.1)

agent = OffPolicyMC(
    env=env,
    epsilon=0.1,
    gamma=1.0,
    init_value=-500,
)

episode_lengths = agent.train(num_episodes=500000)
```

```python
agent.visualize_value_function()
```

```python
agent.test_policy(num_episodes=3)
```
